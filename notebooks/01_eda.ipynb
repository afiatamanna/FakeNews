Load the dataset (place `Fake.csv` and `True.csv` in `data/` or `data/raw/`), inspect class balance, text length, and sample rows.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path("..").resolve() / "src"))

from data_loading import load_raw_data
import pandas as pd

df = load_raw_data()
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("\nLabel distribution:")
print(df["label"].value_counts())
```

```python
df["text_len"] = df["text"].fillna("").astype(str).str.len()
df["word_count"] = df["text"].fillna("").astype(str).str.split().str.len()
df.groupby("label")[["text_len", "word_count"]].describe()
```
